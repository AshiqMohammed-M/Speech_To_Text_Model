```python
from torch import nn
import torch

out_channels = 10
kernal_size = 3
stride = 2

layer = nn.Conv1d(
    out_channels,
    out_channels,
    kernel_size=kernal_size,
    stride=stride,
)
```

```python
batch_size = 4
num_channels = 32
seq_len = 100

input_tensor = torch.randn(batch_size, num_channels, seq_len)
output_tensor = layer(input_tensor)
output_tensor.shape
```
